```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
plt.scatter(iris['petal_length'],
            iris['petal_width'],
            label='petal')
plt.scatter(iris['sepal_length'],
            iris['sepal_width'],
            label='sepal')
plt.legend()
```

```python
sepal_length = iris.groupby('species')['sepal_length'].mean()
sepal_width = iris.groupby('species')['sepal_width'].mean()
petal_length = iris.groupby('species')['petal_length'].mean()
petal_width = iris.groupby('species')['petal_width'].mean()
features = [sepal_length, sepal_width, petal_length, sepal_width]
```

```python
sepal_length
```

```python
fig, ax = plt.subplots()

for feature in features:
    ax.bar(feature.index, feature.values, alpha=0.5)
```

```python
sns.pairplot(data=iris, hue='species')
```

```python
sns.jointplot(data=iris,
              x='sepal_length',
              y='sepal_width',
              kind='hex')
```

```python
plot = sns.FacetGrid(data=iris, col='species', margin_titles=True)
plot.map(plt.hist, 'sepal_length', color='green')

plot = sns.FacetGrid(data=iris, col='species', margin_titles=True)
plot.map(plt.hist, 'sepal_width', color='blue')
```
