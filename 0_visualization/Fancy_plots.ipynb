```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
plt.style.use('ggplot')
```

## Sine & Cosine

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000000)
y_sin = np.sin(x)
y_cos = np.cos(x)


plt.plot(x, y_sin, label='sin')
plt.plot(x, y_cos, label='cos')

plt.title('Sine & Cosine')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()

plt.show()
```

## PDF & CDF

**Probability Density Function & Cumulative Distribution Function**

```python
from pylab import *

# Create some test data
dx = .01
X  = np.arange(-2,2,dx)
Y  = exp(-X**2)

# Normalize the data to a proper PDF
Y /= (dx*Y).sum()

# Compute the CDF
CY = np.cumsum(Y*dx)

# Plot both
plt.plot(X,Y, 'b', label='PDF')
plt.plot(X,CY,'r--', label='CDF')

plt.legend()

plt.show()
```

```python
### Compute the Normal CDF of certain values.
print(stats.norm.cdf(np.array([-1.5, -.5, 0, 1, 1.5, 2, 6])))
```

```python
x = np.linspace(0, 1.5 * np.pi, 10000)
y = np.sin(x ** 2) + np.cos(x ** 2)

fig, axs = plt.subplots(2, 2,
                        subplot_kw=dict(polar=True),
                        figsize=(12, 8))

axs[0,0].plot(x, y)
axs[1,1].scatter(x, y)
```

## Logistic Function
Logistic regression function
$$\sigma(t) = \frac{e^t}{1 + e^t}\$$

```python
x = np.linspace(-6, 6, 10000)
y = (np.exp(x)) / (1 + np.exp(x))
```

```python
plt.plot(x, y, c='b')

plt.title('Logistic Function')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

### Hiperbolic functions

$$y_{1} = cosh(x)$$
$$y_{2} = sinh(x)$$
$$y_{3} = tanh(x)$$

```python
x = np.linspace(-2.5, 2.5, 100000)
x_tan = np.linspace(-5.5, 5.5, 100000)
y1 = np.cosh(x)
y2 = np.sinh(x)
y3 = np.tanh(x_tan)
```

```python
plt.figure(figsize=(8,6))

plt.plot(x, y1, label='cosh(x)')
plt.plot(x, y2, label='sinh(x)')
plt.plot(x_tan, y3, label='tanh(x)')

plt.xlim([-6, 6])

plt.title('Hiperbolic functions')
plt.legend()
```
