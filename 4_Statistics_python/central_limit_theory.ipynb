```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
normal = np.random.normal(10, 5, size=10000)
```

```python
plt.hist(normal, bins=np.linspace(normal.min(), normal.max(), 50));
plt.axvline(normal.mean(), color='orange')
```

```python
uniform = np.random.randint(1, 7, size=10000)
```

```python
plt.hist(uniform);
```

```python
from tqdm import tqdm

means = []
for i in tqdm(range(100000)):
    mean = np.random.randint(1, 7, size=1000).mean()
    means.append(mean)
```

```python
plt.hist(means, bins=np.linspace(np.min(means), np.max(means), 50));
```

Applying log transformation to a positive skewed distribution so it'll end up being a normal distribution.

```python
gamma = np.random.gamma(10, 0.5, 1000)
plt.hist(gamma, bins=25);
plt.axvline(np.mean(gamma), color='orange')
```

```python
plt.hist(np.log1p(gamma), bins=25);
```
